#### seaborn 스타일링
- `set_style`: background color, grid, spine, tick 정의하여 그림의 전반적인 모양 스타일링
- `set_context`: 프레젠테이션이나 보고서와 같은 다양한 매체에 활용할 수 있도록 스타일링

1. `set_style`
    - 내장 태마 활용
        - `darkgrid`(기본값), `whitegrid`, `dark`, `white`, `ticks`

2. `sns.despine(right=T/F, left=T/F, top=T/F, bottom=T/F)`
    - 축선 표시 여부 결정하는 함수
3. `sns.set_context()`
    - 전체 스케일 조정
    - `paper`, `notebook`(기본값), `talk`, `poster`
    - 폰트 크기를 조정하는 파라미터는 별도(`sns.set_context()`안에 `font_scale`)

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_style('dark')
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(right=False, left=False)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_context('talk')
# plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_context('talk', font_scale=.5)
# plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_context('notebook')
# plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

### Distribution Plot

3. box plot
- max, min, mean, first quantile, third quantile
- outlier 발견하기 좋음

```python
sns.set_style('whitegrid')

plt.figure(figsize=(10,6))
sns.boxplot(x=tips['total_bill'])
sns.despine()
plt.show()
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(10,6))
sns.boxplot(x=tips['day'], y=tips['total_bill'])
sns.despine(offset=10)  # x, y축간의 간격을 설정
plt.show()
```

###  지도 생성 방법
- `Map()`메소드에 중심 좌표값을 지정해서 간단하게 생성
- 위도와 경도를 데이터로 지도를 그려줌
- 크롬에서 출력가능

```python
# 초기 map 생성
# 중심좌표 인수 location=[lon,lat]
# 확대비율 정의 인수 zoom_start=13

map_osm = folium.Map(location=[45.5236, -122.6750], zoom_start=13)
map_osm
```

- 지도 스타일 변경

```python
map_osm = folium.Map(location=[45.5236, -122.6750], zoom_start=13, tiles='Stamen Toner') # Stamen Terrain
map_osm
```

- 마커 설정
    - 마커: 특정 위치를 표시하는 표식
    - popup: 마커 클릭시에 나타나는 정보
1. 마커 생성
    - `folium.Marker()`: 인자값 위, 경도 값 리스트, popup문자
2. 마커 생성 후 부착하는 코드
    - `folium.Marker().add_to(지도객체변수)`

```python
map_1 = folium.Map(location=[45.372, -121.6972],
                   zoom_start=11, 
                   tiles='Stamen Terrain')
folium.Marker([45.3288, -121.6625], popup='Hood 산', icon=folium.Icon(icon='cloud')).add_to(map_1)
map_1
```

```python
map_1 = folium.Map(location=[37.579617, 126.977041],
                   zoom_start=13, 
                   tiles='Stamen Terrain')
folium.Marker([37.579617, 126.977041], popup='경복궁', icon=folium.Icon(icon='cloud')).add_to(map_1)
map_1
```

### 단계 구분도
- 데이터를 지도 그림에 반영시켜서 전달하는 그래프
- 보통 folium에 layer로 얹게 된다
- `map` 객체의 `choropleth()` 이용해서 작업
- `folium.LayerControl().add_to(map)`

```python
# 미국 각 주별 실업율을 data로 단계 구분도 작성
```

```python
state_unemployment = './data/02. folium_US_Unemployment_Oct2012.csv'
state_unemp_data = pd.read_csv(state_unemployment)
state_unemp_data.head()
```

```python
state_geo='./data/02. folium_us-states.json'
state_geo
```

```python
map = folium.Map(location=[40,-98], zoom_start=3)

map.choropleth(geo_data=state_geo,
              data=state_unemp_data,
              columns=['State', 'Unemployment'],
              key_on='feature.id',
              fill_color='YlGn',
              legend_name='Unemployment Rate(%)')
folium.LayerControl().add_to(map)
map
```
