## 데이터 프레임 병합
- pandas는 두 개 이상의 데이터프레임을 하나로 합치는 병합(merge)이나 연결(concate)을 지원함

### merge 명령을 사용한 데이터프레임 병합
- `merge`:
    - 두 개의 데이터프레임의 공통 열이나 인덱스를 기준으로
    - 두 개의 데이터프레임을 합친다
    - 이때 기준이 되는 열 데이터 = `key`

#### 형식
- `df.merge(df1)`: 두 df를 병합시켜 준다
- `inner join`(기본): 양쪽에 동일하게 존재하는 키만 표시
- `key`: 기준열을 의미
- 병합방식
    - `inner join`: 양쪽 df에서 모두 키가 존재하는 data만 표시
    - `outer join`: 한쪽에만 키가 존재하면 data를 표시
    - 병합방식을 설정: `how=inner`(생략가능), `how=outer`

```python
df1 = pd.DataFrame({
    '고객번호': [1001,1002,1003,1004,1005,1006,1007],
    '이름': ['둘리','도우너','또치','길동','희동','마이콜','영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001,1001,1005,1006,1008,1001],
    '금액': [10000,20000,15000,5000,100000,30000]
}, columns=['고객번호','금액'])
df2
```

- merge 명령으로 두 df를 병합하는 문법
    - 공통 열: '고객번호'(key)
    - 양쪽에 모두 존재하는 키의 데이터만 보여주는 inner join 방식

```python
df1.head(1)
df2.head(1)
```

```python
df1.merge(df2)
```

```python
pd.merge(df1, df2)
```

- outer join 방식

```python
pd.merge(df1, df2, how='outer')
```

```python
# 1002, 1003, 1004, 1007번 고객은 은행의 회원이지만 예금 X
# 1008번 고객은 예금은 했지만 이름 수집 X
```

```python
pd.merge(df1, df2, how='left')  # 왼쪽 df에 있는 모든 키의 데이터 표시
```

```python
pd.merge(df1, df2, how='right')  # 오른쪽 df에 있는 모든 키의 데이터 표시
```

```python
df1.merge(df2, how='left')
```

```python
df2.merge(df1, how='left')
```

- 동일한 키 값이 있는 경우
    - 키 값이 같은 데이터가 여러개 있는 경우에는 있을 수 있는 모든 경우의 수를 따져서 조합을 만들어낸다.

```python
df1 = pd.DataFrame({
    '품종':['setosa','setosa','virginica','virginica'],
    '꽃잎길이':[1.4,1.3,1.5,1.3],
}, columns=['품종','꽃잎길이'])
df1
```

```python
df2 = pd.DataFrame({
    '품종':['setosa', 'virginica', 'virginica', 'ersicolor'],
    '꽃잎너비':[0.4,0.3,0.5,0.3]
}, columns=['품종', '꽃잎너비'])
df2
```

- df1과 df2를 병합
    - 키 값 `setosa`에 대해
    - df1에는 1.4와 1.3 2개의 데이터가 있고
    - df2에는 0.4라는 1개의 데이터가 있으므로
    - 병합 데이터에는 setosa가 (1.4,0.4)(1.3,0.4)의 두 경우가 표현된다.

```python
pd.merge(df1, df2)
```

- 이름이 같아도 키가 되면 안 되는 열이 있다면 `on` 인수로 기준열 명시해야 함

```python
df1 = pd.DataFrame({
    '고객명':['춘향','춘향','몽룡'],
    '날짜':['2018-01-01', '2018-01-02','2018-01-01'],
    '데이터':[20000,30000,100000]
})
df1
```

```python
df2 = pd.DataFrame({
    '고객명':['춘향','몽룡'],
    '데이터':['여자','남자']
})
df2
```

```python
# 데이터 컬럼은 값의 의미가 다르므로 키로 사용되면 안됨
pd.merge(df1, df2, on='고객명')
```

```python
df1.merge(df2, on='고객명')
```

```python
df1.merge(df2, on='고객명', how='outer')
```

#### 같은 이름의 열이 없는 경우
- `left_on`, `right_on` 인수 사용해서 기준열 명시해야 함

```python
df1 = pd.DataFrame({
    '이름':['영희', '철수', '철수'],
    '성적':[90,80,80]
})
df2 = pd.DataFrame({
    '성명':['영희', '영희', '철수'],
    '성적':[100,80,90]
})
```

```python
df1.head(1)
df2.head(1)
```

```python
pd.merge(df1, df2, left_on='이름', right_on='성명')
# 출력 결과는 양쪽 필드명이 다르기 때문에 기준열로 설정한 두 필드 모두 반환
```

### concat 명령을 사용한 데이터 연결
- `pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None)`
    - `objs`: Series, DF, Panel object
    - `axis=0`: 0은 위아래로 합치기, 1은 왼쪽오른쪽으로 합치기
    - `join='outer'`: `outer`는 합집합, `inner`은 교집합
    - `ignore_index=False`: `False`는 기존 index 유지, `True`는 기존 index 무시
    - `keys=None`: 계층적 index 사용하려면 keys 튜플 입력

```python
s1 = pd.Series([0,1], index=['A','B'])
s2 = pd.Series([2,3,4], index=['A','B','C'])
```

```python
s1
```

```python
s2
```

```python
pd.concat([s1,s2])
```

```python
# concat 연결
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                  index=[0,1,2,3])
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'E':['C4','C5','C6','C7'],
                   'F':['D4','D5','D6','D7']},
                  index=[0,1,2,3])
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'E':['C8','C9','C10','C11'],
                   'F':['D8','D9','D10','D11']},
                  index=[1,2,3,4])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
result = pd.concat([df1, df2, df3])
result
```

```python
result=pd.concat([df1, df2, df3], keys=['x','y','z'])
result
```

```python
result.loc['x'].loc[0,'A']
```

```python
result=pd.concat([df1, df2, df3], join='inner', keys=['x','y','z'])
result
```

```python
result=pd.concat([df1, df2, df3], join='inner')
result
```

```python
result=pd.concat([df1, df2, df3], join='inner', ignore_index=True)
result
```

- `keys=[]`: 상위레벨 인덱스 설정

```python
result = pd.concat([df1, df2, df3], keys=['x','y','z'])
result
```

- 다중 인덱스인 경우 데이터 접근: `.`연산자를 이용한 체인 인덱싱

```python
result.loc['y'].loc[1]
```

```python
result.loc['y'].loc[1:2]
```

#### concat을 이용한 열 병합
- `axis=1` 설정
- `pd.concat([df1, df2], axis=1, join='inner'/'outer')`
- 모든 행 표시하고 해당 행의 데이터가 없는 열의 원소는 NaN으로 표시(기본설정: `join='outer'`)
- 병합하는 데이터프레임에 중복되는 인덱스의 행만 표시(`join='inner'`)

```python
df1 = pd.DataFrame(np.arange(6).reshape(3,2), index=['a','b','c'], columns=['데이터1', '데이터2'])
df2 = pd.DataFrame(5+np.arange(4).reshape(2,2), index=['a','c'], columns=['데이터2', '데이터4'])
```

```python
df1
```

```python
df2
```

```python
# 열방향 결합
# join 방식 생략: outer
pd.concat([df1,df2], axis=1)
```

```python
# 열방향 결합(axis=0 생략가능)
pd.concat([df1,df2], axis=0)
```

```python
pd.concat([df1,df2], axis=1, join='inner')
```

```python
test = pd.concat([df1,df2], axis=1, join='inner')
test.데이터2
```

```python
test = pd.concat([df1,df2], axis=1, join='inner', keys=['x','y'])
test
```

```python
test.x
```

```python
test.x.데이터1
```
