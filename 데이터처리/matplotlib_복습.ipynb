- `color`(c): 선 색깔
- `linewidth`(lw): 선 굵기
- `linestyle`(ls): 선 스타일
- `marker`: 마커의 종류
- `markersize`(ms): 마커의 크기
- `markeredgecolor`(mec): 마커 선 색깔
- `markeredgewidth`(mew): 마커 선 굵기
- `markerfacecolor`(mfc): 마커 내부 색깔

```python
# 스타일을 약자로 표시
plt.figure(figsize=(10,6)) # 단위: inch
plt.plot([10,20,30,40], [1,4,9,16],
         c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.show()
```

```python
plt.figure(figsize=(10,6)) # 단위: inch
plt.plot([10,20,30,40], [1,4,9,16],
         c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')

plt.xlim(0,50)   # X축 범위
plt.ylim(-10,30) # Y축 범위
plt.title("스타일 적용 예")
plt.show()
```

- 하나의 그래프에 여러 데이터를 여러 선으로 표현 가능

```python
t = np.arange(0.,5.,0.2)
plt.figure(figsize=(10,6))
plt.title('라인플롯에서 여러개의 선 그리기')
plt.plot(t,t,'r--') # red, --(dashed)
plt.plot(t, 0.5*t**2, 'bs:') # blue, square marker, dot
plt.plot(t, 0.2*t**3, 'g^-') # green, triangle marker, solid
plt.show()
```

- 위 그래프 코드를 `plot()` 하나로 한번에 표현

```python
plt.plot(t,t,'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
```

```python
plt.figure(figsize=(10,6))
plt.title('라인플롯에서 여러개의 선 그리기: plot 한번 사용')
plt.plot(t,t,'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
plt.show()
```

- 그래프 제목 및 축 레이블 설정
    - `plot.title(data, loc=, pad=, fontsize=)`
        - `loc='right'|'left'|'center'`: `center`가 디폴트
        - `pad=point`: 타이틀과 그래프와의 간격
        - `fontsize`: 제목 폰트 크기
    - `plot.xlabel()`
    - `plot.ylabel()`

```python
x=[10,20,30,40,50,60]
y=[11,15,40,40,20,10]

plt.title('그래프제목', pad=30, fontsize=20)
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
plt.xticks(x, ['10대','20대','30대','40대','50대','60대'])
plt.yticks(y, [y[i] for i in range(6)])
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')

plt.show()
```

- 그래프 Title 폰트 지정
    - `fontdict`
    - 딕셔너리 형식으로 `fontsize`, `fontweight` 지정 가능

```python
plt.title('그래프제목', pad=10, loc='right')

title_font={
    'fontsize': 16,
    'fontweight': 'bold'
}
# fontdict
plt.title('그래프 제목', fontdict=title_font, loc='left', pad=20)
plt.show()
```

- `plot.grid(True)`: 배경 그리드 표시

```python
x=[10,20,30,40,50,60]
y=[11,15,40,40,20,10]

plt.title('그래프제목', pad=30, fontsize=20)
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
plt.xticks(x, ['10대','20대','30대','40대','50대','60대'])
plt.yticks(y, [y[i] for i in range(6)])
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.grid(True)  # tick선 연장해 grid 표현
plt.show()
```

- `plt.subplot()`: 하나의 figure안에 여러개의 플롯을 배열 형태로 표시(그리드 형태의 axes 객체들 생성)
    - `tight_layout(pad=)`: 플롯간 간격

```python
np.random.seed(0)

plt.subplot(221)  # 2행 2열 그리드에서 첫번째 위치
plt.plot(np.random.rand(5))
plt.title('axes1')

plt.subplot(222)  # 2행 2열 그리드에서 2번째 위치
plt.plot(np.random.rand(5))
plt.title('axes2')

plt.subplot(223)  # 2행 2열 그리드에서 3번째 위치
plt.plot(np.random.rand(5))
plt.title('axes3')

plt.subplot(224)  # 2행 2열 그리드에서 4번째 위치
plt.plot(np.random.rand(5))
plt.title('axes4')

plt.tight_layout(pad=1.5)
```

- `plt.subplots(행, 열)`: 여러개의 Axes 객체를 동시에 생성해주는 함수(행렬 형태)
    - 두 개의 반환 값
    - 첫번째는 그래프 객체 전체 이름
    - 두번째는 axes 객체
    - `fig, axes = plt.subplots(2,2)`

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot(np.random.rand(5))
ax[0,0].set_title('axes1')

ax[0,1].plot(np.random.rand(5))
ax[0,1].set_title('axes2')

ax[1,0].plot(np.random.rand(5))
ax[1,0].set_title('axes3')

ax[1,1].plot(np.random.rand(5))
ax[1,1].set_title('axes4')

plt.tight_layout(pad=1.5)
plt.show()
```

### 세로 막대 그래프 그리기
- `bar(x, y, color=[], alpha=)`
    - `alpha`: 투명도 설정

```python
y = [2,3,1,4]
x = np.arange(len(y))
z = [2,3]
s = [0,1]

e = [1,4]
h = [2,3]

xlabel=['가','나','다','라']
plt.figure(figsize=(10,6))
plt.title('Bar Chart')
plt.bar(x,y)

plt.plot(s,z, color='green', linestyle='dashed', marker='o')
plt.plot(h,e, color='blue', linestyle='dashed', marker='o')
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.text(-0.1, 1, r'test')
plt.xlabel('가나다라')
plt.ylabel('빈도수')
plt.show()
x
```

### 가로 막대 그래프 그리기
- `barh(x, y, color=[], alpha=)`

```python
np.random.seed(0)
people=['몽룡', '춘향', '방자', '향단']
y = np.arange(len(people))  # 0, 1, 2, 3
# np.arange(4): 0~ 4 사이의 정수를 순서적으로 추출
# np.random.rand(4): 0~1 사이의 난수를 4개 추출

performance = 3 + 10 * np.random.rand(len(people))

plt.title('Bar Chart')
plt.barh(y, performance, alpha=0.4) # 첫번째 데이터가 y축임!!
plt.yticks(y,people)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
```
