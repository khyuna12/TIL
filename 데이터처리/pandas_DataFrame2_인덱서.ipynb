#### Boolean selection

```python
df.A
```

```python
df.A > 15
```

```python
df.loc[df.A > 15]
```

```python
# 불리언 시리즈를 반환하는 함수 작성
def sel_row(df):
    return df.A > 15
sel_row(df)
```

```python
df.loc[sel_row(df)]
```

## iloc 인덱서(위치 인덱스)

- 위치 정수값은 0부터 시작

```python
df
```

```python
df.iloc[0,1]
```

```python
df.iloc[0:2,1:2]
```

```python
df.iloc[0:2]
```

```python
df.iloc[0:2, 1]
```

```python
df.iloc[2]
```

```python
df.iloc[2:3,1:2]
```

```python
df
```

```python
df.iloc[0:1,-2:]
```

- 원소들을 분류하여 개수를 세는 함수: `value_counts()`

```python
df = pd.DataFrame({'num_legs':[2,4,4,6],
                  'num_wings':[2,0,0,0]},
                 index=['falcon','dog','cat','ant'])
df
```

```python
df.num_legs
```

```python
df.num_legs.value_counts()
```

```python
df.value_counts()
```
