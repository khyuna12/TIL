### 시리즈 연산

- 벡터화 연산(numpy배열 처럼 시리즈도 가능): 집합적 자료형 원소 각각을 독립적으로 계산을 진행하는 방법

```python
pd.Series([1,2,3]) + 4
```

```python
s[(s>250e4) & (s<500e4)]
```

```python
# Boolean Selection
s0 = pd.Series(np.arange(10), np.arange(10)+1)
s0
```

```python
s0 > 5
```

```python
s0[s0>5]
```

```python
s0[s0%2 == 0]
```

```python
s0.index > 5
```

```python
s0[s0.index > 5]
```

```python
s0[(s0>5)&(s0<8)]
```

```python
(s0>=7).sum() # True의 개수 총 합
```

```python
(s0[s0>=7]).sum() # 조건의 결과가 True인 원소들의 합
```

### 딕셔너리로 시리즈 만들기
- `Series({key:value, key:value, ...})`
- index -> key
- 값 -> value

```python
scores={'홍길동':96, '이몽룡': 100, '성춘향':88}
s = pd.Series(scores)
s
```

```python
city = {'서울': 9631482, '부산':3393191, '인천':2632035, '대전':1490158}
s = pd.Series(city, index=city.keys())
s
```

```python
s = pd.Series(city, index=['부산', '인천', '서울', '대전'])
s
```

### Series 함수
- `size`(속성): 개수
- `shape`(속성): 튜플 형태로 shape 반환
- `unique`: 유일한 값만 ndarray로 반환
- `count`: NaN을 제외한 개수 반환
- `mean`: NaN을 제외한 평균
- `value_counts`: NaN을 제외하고 각 값들의 빈도를 반환

```python
s1 = pd.Series([1,1,2,1,2,2,2,1,1,3,3,4,5,5,7,np.NaN])
s1
```

```python
len(s1)
```

```python
s1.size
```

```python
s1.shape
```

```python
s1.unique()  # nan도 하나의 값으로 봄
```

```python
s1.count()  # nan제외
```

```python
a = np.array([2,2,2,2,np.NaN]) # array 타입
a.mean()
```

```python
b = pd.Series(a)
b.mean()
```

```python
s1.value_counts()
```

- pandas 패키지의 `date_range` 함수(날짜 생성)
    - `pd.date_range(start=None, end=None, periods=None, freq='D')`
    - `start`: 시작 날짜, `end`: 끝 날짜, `periods`: 날짜 생성기간, `freq`: 날짜 생성 주기

```python
pd.date_range(start='2018-10-01', end='2018-10-20')
```

```python
pd.date_range(start='2018-10-01', end='2018-10-20', freq = '3D')
```

```python
pd.date_range(start='2018-10-01', end='2018-10-20', freq = 'w')
```

```python
pd.date_range(start='2018-10-01', periods=4, freq = 'w')
```

```python
pd.date_range(start='2018-10-01', periods=4, freq = 'm')
```

```python
pd.date_range(start='2018-10-01', periods=4, freq = 'MS')
```

```python
pd.date_range(start='2018-10-01', periods=12, freq = '2BM')  # 업무일 기준 2개월 간격 월말 주기
```

```python
pd.date_range(start='2018-10-01', periods=4, freq = 'QS') # 분기 시작일 기준
```

```python
pd.date_range(start='2018-10-01', periods=4, freq = 'AS') # 연도 시작일 기준
```

```python
pd.date_range(start='2018-1-20 08:00', periods=10, freq = 'H')
```

```python
pd.date_range(start='2018-1-20 08:00', periods=10, freq = 'BH') # 업무시간 기준 9 to 5 
```

```python
pd.date_range(start='2018-1-20 08:00', periods=10, freq = '30min')
```

```python
pd.date_range(start='2018-1-20 08:00', periods=10, freq = '10S')
```
