### 카테고리 값 세기
- 시리즈 값이 정수, 문자열 등 카테고리 값인 경우에 `시리즈.value_counts()` 메서드를 사용해 각각의 값이 나온 횟수를 셀 수 있음
- `normalize=True`를 사용하면 각 값 및 범주형 데이터의 비율을 계산

```python
np.random.seed(1)
s2 = pd.Series(np.random.randint(6, size=100))
len(s2)
```

```python
s2.value_counts()
```

```python
s2.value_counts(normalize=True)
```

### 행/열 합계
- `df.sum()`
- 열의 합계: `sum(axis=0)`
- 행의 합계: `sum(axis=1)`

```python
np.random.seed(1)
df2 = pd.DataFrame(np.random.randint(10, size=(4,8)))
df2
```

```python
# 각 행의 합계
df2.sum(axis=1)
```

```python
# 긱 열의 합계
df2.sum(axis=0)
```

```python
# axis 인수 생략하면 기본값이 0(각 열의 합계)
df2.sum()
```

```python
df2['total'] = df2[[0,1,2,3,4,5,6,7]].sum(axis=1)
df2
```
