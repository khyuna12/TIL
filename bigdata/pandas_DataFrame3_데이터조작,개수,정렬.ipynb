### 열 또는 행에 동일한 연산 반복 적용: apply() 함수
- `apply()` 함수는 데이터프레임의 행이나 열에 복잡한 연산을 vectorizing 할 수 있게 해주는 함수로 매우 많이 활용되는 함수
- 동일한 연산을 모든열에 혹은 모든 행에 반복 적용하고자 할 때 사용
- `apply(반복적용할 함수, axis=0/1)`
    - 0: 열마다 반복
    - 1: 행마다 반복
    - 생략시 기본값:0

```python
df3 = pd.DataFrame({
    'a':[1,3,4,3,4],
    'b':[2,3,1,4,5],
    'c':[1,5,2,4,4]
})
df3
```

```python
np.sum([1,2,3])  # np.sum() 함수는 전달된 첫번째 인수 데이터들의 합산 결과를 반환함
```

```python
np.sum(df3['a'])
```

```python
np.sum(df3['b'])
```

```python
np.sum(df3['c'])
```

```python
for i in ['a','b','c']:
    print(np.sum(df3[i]))
```

```python
df3.apply(np.sum, 0)  # for문 쓴 거랑 같은 결과
```

#### 데이터값을 카테고리 값으로 변환
- `cut(data, bins, labels)`
    - data: 구간 나눌 실제 값
    - bins: 구간 경계값
    - labels: 카테고리값
- `qcut(data, 구간수, labels)`

```python
ages=[0,0.5,4,6,4,5,2,10,21,23,37,15,38,31,61,20,41,31,100]
```

```python
# data: 구간 나눌 실제 값, bins: 구간 경계값, label: 카테고리 값
data = ages

# 구간 경계값
# 구간 최솟값 < 구간 <= 구간 최댓값
bins=[0,4,18,25,35,60,100]
# 0~4 구간: 영유아 0살 < 영유아 <= 4살

# 각 구간의 이름: labels - 카테고리명
# 순서는 구간(bins)의 순서와 동일해야 함
labels = ['영유아','미성년자','청년','중년','장년','노년']
```

```python
len(data)
```

```python
cats = pd.cut(data, bins, labels=labels)
cats
```

```python
type(cats)
```

```python
cat_list = list(cats)
```

```python
cat_list
```

```python
test = pd.DataFrame({'나이':ages, '연령대':cat_list})
test
```

```python
test['연령대'].value_counts()
```

#### 구간 경계선을 지정하지 않고 데이터 개수가 같도록 지정한 수의 구간으로 분할하기: qcut()
- 형식: `pd.qcut(data, 구간수, labels=[d1,d2, ...])`

```python
# 랜덤정수 20개 생성하고 생성된 정수를 4개의 구간으로 나누기
# 각 구간의 label은 Q1, Q2, Q3, Q4

np.random.seed(2)
data = np.random.randint(20, size=20)
data
```

```python
qcat = pd.qcut(data, 4, labels=['Q1','Q2', 'Q3', 'Q4'])
qcat
```

```python
np.sort(data)
```

```python
pd.value_counts(qcat)
```

```python
df0 = pd.DataFrame(data, columns=['관측수'])
df0['범주'] = qcat
df0
```

### 인덱스 설정 함수
#### 데이터프레임 인덱스 설정을 위해 set_index(), reset_index()
- `set_index()`: 기존 행 인덱스 제거하고 데이터 열 중 하나를 인덱스로 설정
- `reset_index()`: 기존 행 인덱스 제거하고 기본 인덱스로 변경
- 기본 인덱스: 0부터 1씩 증가하는 정수 인덱스

```python
df3 = pd.DataFrame({
    'a':[1,3,4,3,4],
    'b':[2,3,1,4,5],
    'c':[1,5,2,4,4]
})
df3
```

```python
df3.set_index('a') # 원본 반영되지 않음
df3
```

```python
df3.set_index('a', inplace=True)
df3
```

```python
df3.loc[3]
```

```python
# 기본 인덱스로 인덱스 수정하기
df3.reset_index()
df3  # 원본반영안됨
```

```python
# 기존 인덱스 삭제
df3.reset_index(drop=True)
```

```python
df3.reset_index(drop=True, inplace=True)
df3
```

### 리스트 내포 연산
#### 리스트 내포 for문의 일반 문법

- `[표현식(연산식) for 항목 in 반복가능객체 if 조건문]`
- if 조건문 생략가능
- 반복가능객체: 리스트, 튜플, 딕셔너리, range() 등

```python
a = [1,2,3,4]
result = []

for num in a:
    result.append(num*2)
result
```

```python
result = [num*2 for num in a]
result
```

```python
test = []
for i in range(1,5):
    test.append('id_'+str(i))
test
```

```python
test = ['id_'+str(i) for i in range(1,5)]
test
```
