### 데이터프레임 생성

#### 리스트로 데이터프레임 만들기

- `DataFrame([[list1],[list2]])` - 리스트 안에 리스트 형태로 인수 전달(2차원 리스트 형태)
- 각 list는 한 행
- 행의 원소 개수가 다르면 None 값
- index 인수값이 없으면 기본 인덱스(위치 인덱스) 생성됨

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([['a','b','c'],['a','a','g'],['a','a']])
df
```

#### 딕셔너리로 데이터프레임 생성
- dict의 key -> column name

```python
df1 = pd.DataFrame(
{
    'A':[90,80,70],
    'B':[85,98,75],
    'C':[88,99,77],
    'D':[87,89,86]
}, index = [1, 2, 3])
df1
```

```python
data = {
    "2015":[9904312,3448737,2890451,2466052],
    "2010":[9631482,2293191,2632035,2000002],
    "2005":[9762546,3512547,2517680,2456016],
    "2000":[9853972,3655437,2466338,2473990],
    "지역":["수도권","경상권","수도권","경상권"],
    "2010-2015 증가율":[0.0283,0.0163,0.0982,0.0141]
}
df3 = pd.DataFrame(data)
df3
```

```python
columns=["지역", "2000", "2005", "2010", "2015", "2010-2015 증가율"]
index=["서울", "부산", "인천", "대구"]
df3 = pd.DataFrame(data, index=index, columns=columns)
df3
```

#### 시리즈로 데이터프레임 생성
- 각 Series로 인덱스 -> columnname

```python
a = pd.Series([100,200,300], ['a','b','d'])
b = pd.Series([101,201,301], ['a','b','k'])
c = pd.Series([110,210,310], ['a','b','c'])
```

```python
pd.DataFrame([a,b,c], index=[100,101,102])
```

#### csv 데이터로 데이터프레임 생성


```python
train_data = pd.read_csv('./data/train.csv')
train_data.head()
```

- `read_csv` 함수 파라미터
    - `sep`: 각 데이터 값 구별하기 위한 구분자
    - `header`: header 무시할 경우, None 설정
    - `index_col`: index로 사용할 column 설정
    - `usecols`: 실제로 dataframe에 로딩할 columns만 설정

```python
train_data = pd.read_csv('data/train.csv',
                        index_col = 'PassengerId',
                        usecols=['PassengerId','Survived','Name','Sex','Age'])
train_data.head()
```

```python
train_data.columns
```

#### 행 삭제
- `drop()`함수 사용
- `drop(index=[삭제할 행 인덱스])`
- 원본 반영하지 않음(inplace=True 해야함)

```python
df3.drop(index=['광주'], inplace=True)
```

```python
df3
```

```python
df3.drop(columns=['2000'], inplace=True)
df3
```

```python
df3.drop(['대구'], axis=0)
```

```python
df3.drop(['2010-2015 증가율'], axis=1)
```
